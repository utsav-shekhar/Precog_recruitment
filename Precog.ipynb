```python
from datetime import datetime
from datetime import date

def map_function(x, y):
    print(x)
    x, y = x.split('-'), y.split('-')

    a = date(x[2], x[1], x[0])
    b = date(y[2], y[1], y[0])
    return (b - a).days
```

```python
diff = []

for i in  range(len(state_2)):
    try:
        x, y = state_2.iloc[i, 6].split('-'), state_2.iloc[i, 7].split('-')
        a = date(int(x[2]), int(x[1]), int(x[0]))
        b = date(int(y[2]), int(y[1]), int(y[0]))
        diff.append((b - a).days)
    except:
        diff.append(0)
```

```python
diff
```

```python
count = 0
for i in range(len(diff)):
  count += diff[i]

```

```python
print(count/len(diff))
```
